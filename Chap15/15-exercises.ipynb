### 15-1. Cubes:

A number raised to the third power is a cube. Plot the first five cubic numbers, and then plot the first 5000 cubic numbers.

#### a. First five cubic numbers

```python
import matplotlib.pyplot as plt

# ---- First five cubic numbers --------

x_values = [1, 2, 3, 4, 5]
y_values = [x**3 for x in x_values]

plt.scatter(x_values, y_values, s=40)

# Set chart title and label axes.
plt.title("Cubic Numbers", fontsize=18)
plt.xlabel("Value", fontsize=16)
plt.ylabel("Cube", fontsize=16)

plt.tick_params(axis='both', labelsize=14)

plt.show()
```

#### b. First 5000 cubic numbers 

```python
x_values = list(range(5001))
y_values = [x**3 for x in x_values]

plt.scatter(x_values, y_values, edgecolor='none', s=40)

# Set chart title and label axes.
plt.title("Cubic Numbers", fontsize=18)
plt.xlabel("Value", fontsize=16)
plt.ylabel("Cube", fontsize=16)

plt.tick_params(axis='both', labelsize=14)

plt.show()
```

### 15-2. Colored Cubes: 
Apply a colormap to your cubes plot.

```python
x_values = list(range(5001))
y_values = [x**3 for x in x_values]

plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, edgecolor='none', s=40)

# Set chart title and label axes.
plt.title("Cubic Numbers", fontsize=18)
plt.xlabel("Value", fontsize=16)
plt.ylabel("Cube", fontsize=16)

plt.tick_params(axis='both', labelsize=14)

plt.show()
```
