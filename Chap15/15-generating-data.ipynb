Data visualization involves exploring data through visual representations. It's closely associated with data mining, which uses code to explore the patterns and connections in a data set. A data set can be just a small list of numbers that fits in one line of code or many gigabytes of data.

### Plotting a Simple Line Graph

We first import the pyplot module using the alias `plt`so we don't have to type pyplot repeatedly. Then we create a list to hold the squares and then pass it to the `plot()` function, which will try to plot the numbers in a meaningful way. Finally, we use `show()` to display the plot.

```python
# mpl_squares.py
import matplotlib.pyplot as plt

squares = [1, 4, 9, 16, 25]
plt.plot(squares)
plt.show()
```

### Changing the Label Type and Graph Thickness

The `linewidth parameter` controls the thickness of the line that plot() generates. The `title()` function sets a title for the chart. The `fontsize parameter` which appear repeatedly throughout the code, controle the size of the text on the chart. 

The `xlabel()`and `ylabel()`functions allow you to set a title for each of the axes, and the function `tick_params()` styles the tick marks. The arguments shown here affect the tick marks on both the x- and y-axes (axes='both') and set the font size of the tick mark labels to 14.

```python
plt.plot(squares, linewidth=5)

# Set chart title and label axes
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=24)
plt.ylabel("Square of Value", fontsize=24)

# Set size of tick labels
plt.tick_params(axis='both', labelsize=14)

plt.show()
```

### Correcting the Plot

Notice at the end of the graph that the square of 4.0 is shown as 25. So, let's fix that!

```python
input_values = [1, 2, 3, 4, 5]   # - working as x-axis
squares = [1, 4, 9, 16, 25]      # - working as y-axis

plt.plot(input_values, squares, linewidth=5)


# ------ Same as before --------
# Set chart title and label axes
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=24)
plt.ylabel("Square of Value", fontsize=24)

# Set size of tick labels
plt.tick_params(axis='both', labelsize=14)

plt.show()
```

### Plotting and Styling Individual Points with `scatter()`

```python
plt.scatter(2, 4)
plt.show()
```

We call `scatter()` to position the point on the 2D-plane and use the `s` argument to set the size of the dots used to draw the graph. 

```python
plt.scatter(2, 4, s=200)

# Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=24)
plt.ylabel("Square of Value", fontsize=24)

# Set size of tick labels.
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

### Plotting a Series of Points with `scatter()

```python
# plotting a series of points...

x_values = [1, 2, 3, 4, 5]
y_values = [1, 4, 9, 16, 25]

plt.scatter(x_values, y_values, s=100)


# ---------- Same as before ------------
# Set chart title and label axes.
plt.title("Square Numbers", fontsize=24)
plt.xlabel("Value", fontsize=24)
plt.ylabel("Square of Value", fontsize=24)

# Set size of tick labels.
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

### Calculating Data Automatically

We start with a list of x-values conatining the numbers 1 through 1000. Next, a list comprehension generates the y-values by looping through the x-values, squaring each number (x**2), and sotring the results in y_values. We then pass the input and output lists to `scatter()`.

Because this is a large data set, we use a smaller point size and we use the `axis()` function to specify the range of each axis. The `axis()` function requires four values: the minimum and maximum values for the x-axis and the y-axis.

```python
x_values = list(range(1, 10001))
y_values = [x**2 for x in x_values]

plt.scatter(x_values, y_values, s=40)

# Set the range for each axis.
plt.axis([0, 1100, 0, 1100000])

plt.show()
```

###  Removing Outlines from Data Points

__matplotlib__ lets you color points individually in a scatter plot. The default - blue dots with a black outline - works well for plots with a few points. But when plotting many points, tha black outlines can blend together. To remove the outlines around points, pass the argument `edgecolor='none'`when you call `scatter()`

```python
plt.scatter(x_values, y_values, edgecolor='none', s=40)
```

### Defining Custom Colors

To change the color of the points, pass `c` to `scatter()` with the name of a color to use, as shown here:

```python
plt.scatter(x_values, y_values, c='red', edgecolor='none', s=40)
```

`Observation:` You can also define custom colors using the RGB color model. To define a color, pass the `c` as argument a tuple with three decimal values (one each for red, green, and blue), using values between 0 and 1. 

```python
plt.scatter(x_values, y_values, c=(0, 0, 0.8), edgecolor='none', s=40)
```

### Using a Colormap

A colormap is a series of colors in a gradient that moves from a starting to ending color. Colormaps are used in visualizations to emphasize a pattern in the data. For example, you might make low values a light color and high values a darker color.

```python
x_values = list(range(1001))
y_values = [x**2 for x in x_values]

plt.scatter(x_values, y_values, c=y_values, cmap=plt.cm.Blues, edgecolor='none', s=40)
```
